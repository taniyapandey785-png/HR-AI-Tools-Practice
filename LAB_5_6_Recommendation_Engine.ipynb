```python
import pandas as pd

data = {
    "user": ["Amit", "Amit", "Riya", "Riya", "Sohan", "Sohan", "Neha", "Neha", "Kiran", "Kiran"],
    "product": ["AI Book", "Python Book", "AI Book", "Marketing Book", "Python Book", "Finance Book", "AI Book", "Finance Book", "Marketing Book", "AI Book"],
    "rating": [5, 4, 4, 5, 5, 3, 4, 4, 5, 5]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

print("\nAverage Rating by Product:")
product_ratings = df.groupby("product")["rating"].mean().sort_values(ascending=False)
print(product_ratings)

recommendations = product_ratings.reset_index()
recommendations.columns = ["Product", "Average Rating"]

print("\nTop Recommended Products:")
print(recommendations.head(3))
```
