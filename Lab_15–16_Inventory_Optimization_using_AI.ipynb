```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

# Step 1: Generate dataset
np.random.seed(42)
n = 30

data = {
    'daily_sales': np.random.randint(10, 100, n),
    'stock_available': np.random.randint(20, 150, n),
    'lead_time': np.random.randint(1, 10, n),
    'demand_trend': np.random.randint(1, 4, n)
}

df = pd.DataFrame(data)

# Step 2: Create target
df['reorder'] = (
    (df['stock_available'] < df['daily_sales'] * df['lead_time']) |
    ((df['demand_trend'] == 3) & (df['stock_available'] < 80))
).astype(int)

print("Dataset:\n", df.head())

# Step 3: Features and target
X = df[['daily_sales', 'stock_available', 'lead_time', 'demand_trend']]
y = df['reorder']

# Step 4: Train model
model = DecisionTreeClassifier(max_depth=3)
model.fit(X, y)

# Step 5: Predict
sample = [[60, 50, 5, 3]]
prediction = model.predict(sample)

print("\nPrediction:", prediction[0])

if prediction[0] == 1:
    print("Reorder Required")
else:
    print("Stock is Sufficient")
```
