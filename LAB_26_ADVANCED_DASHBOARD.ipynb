```python
# Cell 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Cell 2: Create a sample HR dataset
np.random.seed(42)

n = 100

departments = ['HR', 'Sales', 'IT', 'Finance', 'Operations']
genders = ['Male', 'Female']
recruitment_sources = ['LinkedIn', 'Referral', 'Job Portal', 'Campus']
attrition_options = ['Yes', 'No']
engagement_levels = ['Low', 'Medium', 'High']
promotion_status = ['Promoted', 'Not Promoted']

df = pd.DataFrame({
   'employee_id': [f'E{str(i).zfill(3)}' for i in range(1, n+1)],
   'department': np.random.choice(departments, n),
   'gender': np.random.choice(genders, n),
   'age': np.random.randint(22, 50, n),
   'experience_years': np.random.randint(1, 20, n),
   'monthly_salary': np.random.randint(25000, 120000, n),
   'satisfaction_score': np.random.randint(3, 10, n),
   'training_hours': np.random.randint(5, 60, n),
   'attendance_percent': np.random.randint(70, 100, n),
   'performance_rating': np.random.randint(1, 6, n),
   'overtime_hours': np.random.randint(0, 25, n),
   'attrition': np.random.choice(attrition_options, n, p=[0.25, 0.75]),
   'recruitment_source': np.random.choice(recruitment_sources, n),
   'engagement_level': np.random.choice(engagement_levels, n),
   'promotion_status': np.random.choice(promotion_status, n)
})

df.head()
# Cell 3: Basic summary
print("Dataset Shape:", df.shape)
print("\nColumn Names:")
print(df.columns)
# Cell 4: KPI calculations
total_employees = df.shape[0]
avg_satisfaction = df['satisfaction_score'].mean()
avg_attendance = df['attendance_percent'].mean()
attrition_count = (df['attrition'] == 'Yes').sum()
avg_performance = df['performance_rating'].mean()

print("Total Employees:", total_employees)
print("Average Satisfaction Score:", round(avg_satisfaction, 2))
print("Average Attendance:", round(avg_attendance, 2))
print("Attrition Count:", attrition_count)
print("Average Performance Rating:", round(avg_performance, 2))
# Cell 5: Attrition by department
attrition_by_dept = df[df['attrition'] == 'Yes']['department'].value_counts()
attrition_by_dept.plot(kind='bar')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Cell 6: Average satisfaction score by department
satisfaction_by_dept = df.groupby('department')['satisfaction_score'].mean()
satisfaction_by_dept.plot(kind='bar')
plt.title('Average Satisfaction Score by Department')
plt.xlabel('Department')
plt.ylabel('Satisfaction Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Cell 7: Attendance percentage by department
attendance_by_dept = df.groupby('department')['attendance_percent'].mean()
attendance_by_dept.plot(kind='bar')
plt.title('Average Attendance Percentage by Department')
plt.xlabel('Department')
plt.ylabel('Attendance Percent')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Cell 8: Performance rating distribution
df['performance_rating'].plot(kind='hist', bins=5)
plt.title('Performance Rating Distribution')
plt.xlabel('Performance Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
# Cell 9: Overtime hours by attrition
overtime_attrition = df.groupby('attrition')['overtime_hours'].mean()
overtime_attrition.plot(kind='bar')
plt.title('Average Overtime Hours by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Overtime Hours')
plt.tight_layout()
plt.show()
# Cell 10: Recruitment source breakdown
df['recruitment_source'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Recruitment Source Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
# Cell 11: Business insights
print("Business Insights:")
print("1. Departments with higher attrition may require HR intervention.")
print("2. Lower satisfaction may indicate engagement problems.")
print("3. Higher overtime may be linked with employee burnout.")
print("4. Recruitment sources can be evaluated based on quality of hires.")
```
