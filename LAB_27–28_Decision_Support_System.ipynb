```python
# Cell 1: Import Libraries
# Import required libraries
import pandas as pd
import numpy as np


# Cell 2: Create Synthetic HR Dataset
# Set random seed for reproducibility
np.random.seed(42)

# Number of employees
n = 100

# Create dataset
df = pd.DataFrame({
    'employee_id': [f'E{str(i).zfill(3)}' for i in range(1, n+1)],
    'department': np.random.choice(['HR', 'Sales', 'IT', 'Finance', 'Operations'], n),
    'age': np.random.randint(22, 50, n),
    'experience_years': np.random.randint(1, 20, n),
    'satisfaction_score': np.random.randint(3, 10, n),
    'attendance_percent': np.random.randint(65, 100, n),
    'overtime_hours': np.random.randint(0, 25, n),
    'performance_rating': np.random.randint(1, 6, n)
})

# Display first few rows
df.head()


# Cell 3: Understand Dataset Structure
# Check dataset information
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns)

# Summary statistics
df.describe()


# Cell 4: Create Risk Classification Logic (Core DSS)
# Function to classify employee risk
def classify_risk(row):
    if row['satisfaction_score'] < 5 and row['overtime_hours'] > 15:
        return 'High Risk'
    elif row['satisfaction_score'] < 6 or row['attendance_percent'] < 75:
        return 'Medium Risk'
    else:
        return 'Low Risk'

# Apply classification
df['risk_level'] = df.apply(classify_risk, axis=1)

# View results
df[['employee_id', 'satisfaction_score', 'attendance_percent', 'overtime_hours', 'risk_level']].head()


# Cell 5: Create Decision Recommendation System
# Function to recommend actions
def recommend_action(row):
    if row['risk_level'] == 'High Risk':
        return 'Immediate HR intervention, workload reduction, counseling'
    elif row['risk_level'] == 'Medium Risk':
        return 'Monitor employee, provide training and engagement programs'
    else:
        return 'Maintain current strategy and reward performance'

# Apply recommendations
df['recommended_action'] = df.apply(recommend_action, axis=1)

# Display results
df[['employee_id', 'risk_level', 'recommended_action']].head()


# Cell 6: Risk Distribution Analysis
# Count employees in each risk category
risk_counts = df['risk_level'].value_counts()

print("Risk Distribution:")
print(risk_counts)


# Cell 7: Department-wise Risk Analysis
# Group by department and risk
dept_risk = df.groupby(['department', 'risk_level']).size().unstack(fill_value=0)

print("Department-wise Risk Distribution:")
print(dept_risk)


# Cell 8: High-Risk Employee Identification
# Filter high-risk employees
high_risk_employees = df[df['risk_level'] == 'High Risk']

print("High Risk Employees:")
high_risk_employees[['employee_id', 'department', 'satisfaction_score', 'overtime_hours']]


# Cell 9: Business Insights Generation
# Generate simple business insights
print("Business Insights:")

high_risk_count = len(df[df['risk_level'] == 'High Risk'])

if high_risk_count > 20:
    print("- High number of employees at risk. Immediate HR action required.")
else:
    print("- Risk levels are under control.")

low_satisfaction = df[df['satisfaction_score'] < 5].shape[0]
print(f"- Employees with low satisfaction: {low_satisfaction}")

high_overtime = df[df['overtime_hours'] > 15].shape[0]
print(f"- Employees with high overtime: {high_overtime}")


# Cell 10: Export Results (Optional)
# Save DSS output to CSV
df.to_csv("employee_decision_support_output.csv", index=False)

print("DSS output file saved successfully.")
```
