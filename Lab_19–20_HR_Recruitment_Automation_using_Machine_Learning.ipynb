```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

# Step 1: Generate dataset (50 records)
np.random.seed(42)
n = 50

data = {
    'experience_years': np.random.randint(0, 11, n),
    'test_score': np.random.randint(40, 100, n),
    'interview_score': np.random.randint(40, 100, n),
    'skill_match': np.random.randint(1, 4, n)
}

df = pd.DataFrame(data)

# Step 2: Create target variable using recruitment rules
df['selected'] = (
    ((df['experience_years'] >= 3) & (df['test_score'] >= 60) & (df['interview_score'] >= 60)) |
    ((df['skill_match'] == 3) & (df['interview_score'] >= 70))
).astype(int)

# Display dataset preview
print("Dataset Preview:\n")
print(df.head())

print("\nTotal Records:", len(df))

# Step 3: Define features and target
X = df[['experience_years', 'test_score', 'interview_score', 'skill_match']]
y = df['selected']

# Step 4: Train Decision Tree model
model = DecisionTreeClassifier(max_depth=4, random_state=42)
model.fit(X, y)

# Step 5: Predict for a new candidate
sample = [[4, 75, 80, 3]]  # experience, test score, interview score, skill match

prediction = model.predict(sample)

# Step 6: Output interpretation
print("\nPrediction Result:", prediction[0])

if prediction[0] == 1:
    print("Candidate Likely to be Selected")
else:
    print("Candidate Likely to be Rejected")
```
