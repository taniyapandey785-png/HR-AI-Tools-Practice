```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

# Step 1: Generate dataset (50 records)
np.random.seed(42)
n = 50

data = {
    'distance_km': np.random.randint(5, 200, n),
    'traffic_level': np.random.randint(1, 4, n),
    'weather_condition': np.random.randint(1, 4, n),
    'vehicle_capacity': np.random.randint(50, 100, n)
}

df = pd.DataFrame(data)

# Step 2: Create target variable using business rules
df['delay'] = (
    ((df['distance_km'] > 100) & (df['traffic_level'] == 3)) |
    (df['weather_condition'] == 3) |
    (df['vehicle_capacity'] > 90)
).astype(int)

# Display dataset preview
print("Dataset Preview:\n")
print(df.head())

print("\nTotal Records:", len(df))

# Step 3: Define features and target
X = df[['distance_km', 'traffic_level', 'weather_condition', 'vehicle_capacity']]
y = df['delay']

# Step 4: Train Decision Tree model
model = DecisionTreeClassifier(max_depth=4, random_state=42)
model.fit(X, y)

# Step 5: Predict for a new delivery scenario
sample = [[120, 3, 2, 85]]  # distance, traffic, weather, capacity

prediction = model.predict(sample)

# Step 6: Output interpretation
print("\nPrediction Result:", prediction[0])

if prediction[0] == 1:
    print("Delivery Delay Expected")
else:
    print("Delivery Expected On Time")
```
